## Adjective and Adverbs
- Quelque = Some
    - Il me reste quelque temps / *I have a little time left* 
    - Je voudrais trouver quelque poste. / *I would like to find some job*
    - J'ai fait quelque cent mètres avant d'arriver à une route pavée / *I have walked about one hundred meters before arriving at a paved road*
- Quelque..N..que = No matter + N
    - Quelques peurs que **vous ayez**, conquérez-les / *No matter fears you have, conquer them!*
    - Quelque difficiles que **soient les problèmes**, persérvez. *OR*
    - Quelque difficiles que **les problèmes soient**, persérvez / *No matter how difficult the problems are, preserve!*
    - Quelque bêtement que vous agissiez, je vous aime toujours. / *No matter stupid you act, I always like you*
- Quelconque = Only
    - Ce n'est pas une peinture quelconque. C'est une peinture de Picasso! / *It's not only the painting. It's the painting of Picasso!*
- Simping the Reported Speech
    - Je ne sais quelle **personne a dit "Il n'y a pas de feu"**
    - Je ne sais quelle **les**
- Tel / Tels / Telle / Telles / Comme = Like + N
    - Tel père, tel fils. / *Like father, 
    - Elle est partie telle une flèche. / *She left like an arrow*

    - Avez-vous jamais vu une si belle peinture?. / *Have u ever seen such a beautiful paining?*

## Prepositions
- Certain Prepositions
    - au / à la / aux = to
    - du / de la / des = from
    - dans le / dans la / en = in
    - Il va à la maison / *He goes to the home*
    - Il va ~~à le~~ au bureau / *He goes to the office*
    - Il habite ici depuis deux ans / *He has been living here for two years*
    - Il parle d’une voix basse. / *He speaks in a low voice*

- Synonyms
    - Il travaille **pour** gagner de l’argent = Il travaille **afin de** gagner de l’argent

- Useful verbs
    - Monter vs Descendre
        - Monter dans un autobus / une voiture
        - Descendre d'un autobus / d'une voiture
    - Embarquer vs Débarquer
        - Embarquer / Debarquer dans un bateau
    - Une maison à vendre / à louer / *... for sale / for rent*
- Time and Position
    - Avant vs Après (Before vs After)
    - Devant vs Derrière (Depan vs Belakang)
- À vs de
    - Un verre à eau / *A glass for water*
    - Un verre d’eau / *A glass of water*

    - Il me parle avec faiblesse. / *...with weak*

    - Il me regarde avec fureur. / *...with fear*
    - Il le fait avec adresse. / *...with skill*
